```python
def prep(file, percent=0.149):
    df_pred = pd.read_csv(file)
#    df_pred2 = pd.read_csv(f2)
#    df_pred = df_pred1.append(df_pred2)
    del df_pred['RowId']
    df_pred.set_index(df_test.EventId, inplace=True)
    pct_ix = int(len(df_pred.index) - (len(df_pred.index) * percent))
    cutoff = df_pred['Prediction'].ix[df_pred.Prediction.order().index].iloc[pct_ix]
    df_pred['Class'] = df_pred['Prediction'] > cutoff
    df_pred.Class = ['s' if x else 'b' for x in df_pred.Class]
    # check that the cutoff is correct
    sss = (df_pred.Class == 's').astype(int).sum().astype(float) / len(df_pred.index)
    print "The cutoff was: " + str(cutoff) + ' at ' + str(sss) + ' %'
    df_rank = pd.DataFrame(data=[x for x in range(1,len(df_pred)+1)], 
                       index=df_pred.Prediction.order().index, 
                       columns=['RankOrder'])
    df_pred['RankOrder'] = df_rank.RankOrder
    return df_pred
```

```python
def blend(df1, df2, percent=0.149, mult=0.5):
    s_blend = (mult * df1.RankOrder) + ((1 - mult) * df2.RankOrder)
    df_rank = pd.DataFrame(data=[x for x in range(1,len(s_blend)+1)], index=s_blend.order().index, columns=['RankOrder'])
    df_rank.sort_index(inplace=True)
    pct_ix = int(len(df_rank.index) - (len(df_rank.index) * percent))
    cutoff = df_rank['RankOrder'].ix[df_rank.RankOrder.order().index].iloc[pct_ix]
    df_rank['Class'] = df_rank['RankOrder'] > cutoff
    df_rank.Class = ['s' if x else 'b' for x in df_rank.Class]
    return df_rank
```
