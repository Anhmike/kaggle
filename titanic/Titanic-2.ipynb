```python
csv_dir = '/Users/Paul/Desktop/titanic/'
```

```python
df_train = pd.read_csv(csv_dir + 'train.csv')
```

```python
df_train[:20]
```

```python
# extract the deck of the cabin is on
df_train.loc['Deck'] = df_train.Cabin.apply(lambda x: x[0] if not pd.isnull(x) else x)
```

```python
# extract the # of the cabin, if multiple, just take the first one
for i in df_train.index:
    if not pd.isnull(df_train.Cabin[i]):
        df_train.loc['Num'] = df_train.Cabin[i][1:].split(' ')[0]
```

```python
# extract the last name 
df_train['LastName'] = df_train['Name'].apply(lambda x: x.split(',')[0])
```

```python
# extract title
df_train['Title'] = df_train['Name'].apply(lambda x: x.split(' ')[1] if not pd.isnull(x) else x)
```

```python
l_titles = ["Dr.", "Master.", "Mrs.", "Miss.", "Mr."]
df_train.Title = df_train.Title.apply(lambda x: None if x not in l_titles else x)
```
