```python
# extract the last name 
df['LastName'] = df['Name'].apply(lambda x: x.split(',')[0] if not pd.isnull(x) else x)
```

```python
# extract title, 
df['Title'] = df.Name.apply(lambda x: x.split(',')[1].split(' ')[1])
```

```python
# drop all the non titles
l_titles = ["Dr.", "Master.", "Mrs.", "Miss.", "Mr."]
l_nobile = ["Capt.", "Col.", "Don", "Jonkheer", "Lady", "Major", "Rev", "Sir"]
df['Title'] = df.Title.apply(lambda x: None if x not in l_titles else x)
```

```python
# find the average age by title
dict_age = {title :0 for title in l_titles}
for title in l_titles:
    dict_age[title] = int(df[df.Title == title].Age.mean())
dict_age
```

```python
# impute the average age
for i in df.index:
    if pd.isnull(df.loc[i].Age):
        if not pd.isnull(df.loc[i].Title):
            df.loc[i, 'Age'] = dict_age[df.loc[i].Title]
```
